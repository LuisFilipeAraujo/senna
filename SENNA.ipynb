```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
!curl -o 91DE_122-5_crop-4x3.WEBP https://mclaren.scene7.com/is/image/mclaren/91DE_122-5:crop-4x3?wid=1920&hei=1440
img = cv2.imread('91DE_122-5_crop-4x3.WEBP')
plt.imshow(img)
RGB_im = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_im)
```

```python
dimensions = img.shape
dimensions
```

### **Separando bandas RGB**

```python
r = RGB_im.copy()
r = r[:, :, 0]
g = RGB_im.copy()
g = g[:, :, 1]
b = RGB_im.copy()
b = b[:, :, 2]
```

```python
fig = plt.figure(figsize=(30,20))
ax1 = fig.add_subplot (131) 
plt.xticks([]), plt.yticks([])
plt.imshow(r,cmap='Reds_r') 
plt.title("Mapa de Cor Vermelho")

ax2 = fig.add_subplot (132) 
plt.xticks ([]), plt.yticks([]) 
plt.imshow (g, cmap='Greens_r') 
plt.title("Mapa de Cor Verde")


ax3 = fig.add_subplot (133)
plt.xticks ([]), plt.yticks([]) 
plt.imshow (b, cmap='Blues_r') 
plt.title("Mapa de Cor Azul")

plt.show()
```

### **AMOSTRAGEM**

```python
dimensions_rgb = img.shape
print('Dimensões originais', dimensions_rgb)

n = 5
img_red5 = RGB_im[::n, ::n]
dimensions5 = img_red5.shape
print('Dimensão Imagem n5', dimensions5)

n = 10
img_red10 = RGB_im[::n, ::n]
dimensions10 = img_red10.shape
print('Dimensão Imagem n10', dimensions10)

n = 30
img_red30 = RGB_im[::n, ::n]
dimensions30 = img_red30.shape
print('Dimensão Imagem n30', dimensions30)

```

```python
fig = plt.figure(figsize=(30,20))
ax0 = fig.add_subplot (141) 
plt.imshow(RGB_im)
plt.title("Amostragem 900x1920")
plt.xticks ([]), plt.yticks([])

ax1 = fig.add_subplot (142) 
plt.imshow(img_red5)
plt.title("Amostragem 180x384")
plt.xticks ([]), plt.yticks([])

ax2 = fig.add_subplot (143) 
plt.imshow(img_red10)
plt.title("Amostragem 90x192")
plt.xticks ([]), plt.yticks([])

ax3 = fig.add_subplot (144) 
plt.imshow(img_red30)
plt.title("Amostragem 30x64")
plt.xticks ([]), plt.yticks([])
plt.show()
```

### **Aumentando o tamanho (Resolução)**

```python
dimensions_rgb = img_red10.shape
print('Dimensão Imagem original', dimensions_rgb)

m=5
img_aum5 = np.repeat(img_red5, m, axis=0)
img_aum5 = np.repeat(img_aum5, m, axis=1)
dimensions5 = img_aum5.shape
print('Dimensão Imagem m5', dimensions5)

m=10
img_aum10 = np.repeat(img_red10, m, axis=0)
img_aum10 = np.repeat(img_aum10, m, axis=1)
dimensions10 = img_aum10.shape
print('Dimensão Imagem m10', dimensions10)

m=30
img_aum30 = np.repeat(img_red30, m, axis=0)
img_aum30 = np.repeat(img_aum30, m, axis=1)
dimensions30 = img_aum30.shape
print('Dimensão Imagem m30', dimensions30)
```

```python
fig = plt.figure(figsize=(30,20))
ax0 = fig.add_subplot (141) 
plt.imshow(img_red10)
plt.xticks([]), plt.yticks([])

ax1 = fig.add_subplot (142) 
plt.imshow(img_aum5)
plt.xticks([]), plt.yticks([])

ax2 = fig.add_subplot (143) 
plt.imshow(img_aum10)
plt.xticks([]), plt.yticks([])

ax3 = fig.add_subplot (144) 
plt.imshow(img_aum30)
plt.xticks([]), plt.yticks([])



plt.show()
```

### **QUANTIZAÇÃO**

```python
r=128 # 2 TONS DE CINZA
img_quant128 = np.uint8(RGB_im / r) *r

r=64 # 4 TONS DE CINZA
img_quant64 = np.uint8(RGB_im / r) *r

r=32 # 8 TONS DE CINZA
img_quant32 = np.uint8(RGB_im / r) *r

r=16 # 16 TONS DE CINZA
img_quant16 = np.uint8(RGB_im / r) *r

r=9 # 28,5 TONS DE CINZA
img_quant9 = np.uint8(RGB_im / r) *r

r=4 # 64 TONS DE CINZA
img_quant4 = np.uint8(RGB_im / r) *r

```
