## **ROTACIONAR**

```python
Altura = (RGB_im.shape[0])
Largura = (RGB_im.shape[1])
Canais = (RGB_im.shape[2])
```

```python
print("Altura (height): %d pixels" %Altura)  
print("Largura (width): %d pixels" %Largura) 
print("Canais (channels): %d" %Canais)
```

```python
centro = (Largura // 2, Altura // 2)
centro
```

```python
M = cv2.getRotationMatrix2D(centro, 45, 1.0) #30 graus 
```

```python
img_rotacionada = cv2.warpAffine(RGB_im, M, (Largura, Altura))
```

```python
plt.imshow(img_rotacionada)
plt.title("Imagem Rotacionada")
plt.show()
```
